Chelsea has more England players than any other Country

```python
col=['Age', 'Potential', 'Overall', 'Marking', 'Finishing']
def plot_density_hist(df, cols, bins = 10, hist = False):
    for col in cols:
        sns.set_style("whitegrid")
        sns.distplot(df[col], bins = bins, rug=True, hist = hist)
        plt.title('Histogram of ' + col) # Give the plot a main title
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel('Frequecy')# Set text for y axis
        plt.show()
        
plot_density_hist(Chelsea, col)
```

The mean of the Age is round 25years in Chelsea. The Finishing plot is rightly skewed which means a good percentage above average of chelsea players are good finishers. The potential plot for Chelsea players follows a normal distribution and with the mean of 82.6.
